```python
df_marital_and_education = df[['Income','marital_Divorced', 'marital_Married',
'marital_Single', 'marital_Together', 'marital_Widow', 
'education_Basic', 'education_Graduation',
'education_Master', 'education_PhD', 'MntTotal']]
```

Get all one hot encoded columns and income and Amount total for further analysis

```python
df_marital_and_education
```

```python
df_marital_and_education_melt = df_marital_and_education.melt(id_vars=['Income','MntTotal'],value_vars=['marital_Divorced', 'marital_Married',
'marital_Single', 'marital_Together', 'marital_Widow', 
'education_Basic', 'education_Graduation',
'education_Master', 'education_PhD', 'MntTotal'],var_name='Situation',value_name='Yes&NO')
```

```python
df_marital_and_education_melt['Yes&NO'] = ['Yes' if x == 1 else 'No' for x in df_marital_and_education_melt['Yes&NO']]
df_marital_and_education_melt
```

```python
df_marital_and_education_melt['SituationCombined'] = df_marital_and_education_melt['Situation']+'_'+df_marital_and_education_melt['Yes&NO']
df_marital_and_education_melt['Count'] = 1
df_marital_and_education_melt
```

```python
df_marital_and_education_group = df_marital_and_education_melt.groupby(by='SituationCombined')\
.agg({'MntTotal':'sum','Income':'sum','Count':'sum'}).reset_index()
```

```python
df_marital_and_education_group['avg_Mnt'], df_marital_and_education_group['avg_Income'] = \
df_marital_and_education_group['MntTotal']/df_marital_and_education_group['Count'],\
df_marital_and_education_group['Income']/df_marital_and_education_group['Count']

df_marital_and_education_group
```

```python
plt.figure(figsize=(12,6))
sns.barplot(df_marital_and_education_group.sort_values(by='avg_Mnt',ascending=False)\
            ,x='SituationCombined',y='avg_Mnt',hue='SituationCombined')
plt.xticks(rotation=90)
```

We could see marital_widow and education level PHD tend to spend more than others and education with baisc only spends less

Futher machine learning use case research such as using Kmeans to cluster different customer profile.

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df_cluster_scaled = scaler.fit_transform(df_cluster)
```

```python
kmean = KMeans(n_clusters=3, random_state=42)
```

```python
kmean.fit(df_cluster_scaled)
```

```python
predictions  = kmean.predict(df_cluster_scaled)
```

```python
df_cluster['prediction'] = predictions
```

```python
df_cluster
```

```python
sns.scatterplot(df_cluster, x='Income', y='MntTotal',hue='prediction')
```

```python
g = sns.FacetGrid(df_cluster,col='prediction')
g.map(sns.scatterplot,'Income', 'MntTotal')
```
